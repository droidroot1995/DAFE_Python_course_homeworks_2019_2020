```python
import sys
from PyQt5.QtCore import Qt
from collections import namedtuple
from PyQt5.QtWidgets import *

res = [0,0,0]
sign1 = ''
sign2 = ''
pr1 = 0
pr2 = 0
text = ''
tb = ''
f1 = 0

def w1():
    global res, text, tb
    res[0] = res[0]*10 + 1
    text = text + '1'
    tb.setText(text)
    
def w2():
    global res, text, tb
    res[0] = res[0]*10 + 2
    text = text + '2'
    tb.setText(text)
    
def w3():
    global res, text, tb
    res[0] = res[0]*10 + 3
    text = text + '3'
    tb.setText(text)
    
def w4():
    global res, text, tb
    res[0] = res[0]*10 + 4
    text = text + '4'
    tb.setText(text)
    
def w5():
    global res, text, tb
    res[0] = res[0]*10 + 5
    text = text + '5'
    tb.setText(text)
    
def w6():
    global res, text, tb
    res[0] = res[0]*10 + 6
    text = text + '6'
    tb.setText(text)
    
def w7():
    global res, text, tb
    res[0] = res[0]*10 + 7
    text = text + '7'
    tb.setText(text)
    
def w8():
    global res, text, tb
    res[0] = res[0]*10 + 8
    text = text + '8'
    tb.setText(text)
    
def w9():
    global res, text, tb
    res[0] = res[0]*10 + 9
    text = text + '9'
    tb.setText(text)
    
def w0():
    global res, text, tb
    res[0] = res[0]*10 + 0
    text = text + '0'
    tb.setText(text)
    
def error():
    global sign1, sign2, pr1, pr2, res, text, f1, tb
    res = [0,0,0]
    sign1 = ''
    sign2 = ''
    pr1 = 0
    pr2 = 0
    tb.setText(text)
    text = ''
    f1 = 0
    
def axaxa():
    global sign1, sign2, pr1, pr2, res, text
    if (sign2 == '*'):
        res[2] = res[1]*res[2]
    elif (res[1] == 0):
        text = 'Error'
    else:
        res[2] = res[2]/res[1]
    res[1] = res[0]
    pr2 = pr1
    sign2 = sign1
            
    
def yxyxy():    
    global sign1, sign2, res, text
    if (sign1 == '*'):
        res[1] = res[1]*res[0]
    elif (res[0] == 0):
        text = 'Error'
    else:
        res[1] = res[1]/res[0]
    
def oxoxo():    
    global sign1, sign2, pr1, pr2, res
    if (sign2 == '+'):
        res[2] = res[1]+res[2]
    else:
        res[2] = res[2]-res[1]
    res[1] = res[0]
    pr2 = pr1
    sign2 = sign1
    
    
def ixixi():
    global sign1, sign2, pr1, pr2, res, f1
    sign2 = sign1
    pr2 = pr1
    res[2] = res[1]
    res[1] = res[0]
    
def exexe():
    global sign1, sign2, pr1, pr2, res, f1
    if (pr2 > pr1):
        axaxa()
    elif (pr2 < pr1):
        yxyxy()
    elif (pr1 == 1):
        oxoxo()
    else: axaxa()
    f1 = 1
                    
    
def w_plus():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (f1 == 1):
        ixixi()
    elif (f1 == 2):
        exexe()
    else: 
        res[1] = res[0]
    f1 = f1 + 1
    sign1 = '+'
    pr1 = 1
    res[0] = 0
    if (text == 'Error'):
        error()
    else:
        text = text + sign1
        tb.setText(text) 
            
    
def w_minus():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (f1 == 1):
        ixixi()
    elif (f1 == 2):
        exexe()
    else: 
        res[1] = res[0]
    f1 = f1 + 1
    sign1 = '-'
    pr1 = 1
    res[0] = 0
    if (text == 'Error'):
        error()
    else:
        text = text + sign1
        tb.setText(text) 
    
def w_multiply():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (f1 == 1):
        ixixi()
    elif (f1 == 2):
        exexe()
    else: 
        res[1] = res[0]
    f1 = f1 + 1
    sign1 = '*'
    pr1 = 2
    res[0] = 0
    if (text == 'Error'):
        error()
    else:
        text = text + sign1
        tb.setText(text)    
    
def w_divide():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (f1 == 1):
        ixixi()
    elif (f1 == 2):
        exexe()
    else: 
        res[1] = res[0]
    f1 = f1 + 1
    sign1 = '/'
    pr1 = 2
    res[0] = 0
    if (text == 'Error'):
        error()
    else:
        text = text + sign1
        tb.setText(text) 
    
def w_clear():
    global sign1, sign2, pr1, pr2, res, text, f1
    res = [0,0,0]
    sign1 = ''
    sign2 = ''
    pr1 = 0
    pr2 = 0
    text = ''
    f1 = 0
    tb.setText(text)
    
def clear():
    global sign1, sign2, pr1, pr2, res, text, f1  
    res = [0,0,0]
    sign1 = ''
    sign2 = ''
    pr1 = 0
    pr2 = 0
    text = ''
    f1 = 0

    
    
def su():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (sign2 == '+'):
        res[2] = res[1]+res[2]
    elif (sign2 == '-'):
        res[2] = res[2]-res[1]
    elif (sign2 == '*'):
        res[2] = res[1]*res[2]
    elif (res[1] == 0):
        text = 'Error'
    else:
        res[2] = res[2]/res[1]
    if (text == 'Error'):
        error()
    else:
        text = f'{res[2]}'
        tb.setText(text)
    
def f1S():
    global sign1, sign2, pr1, pr2, res, text, f1
    ixixi()
    su()
    
    
def f2S():
    global sign1, sign2, pr1, pr2, res, text, f1
    exexe()
    su()
    
    
def w_equals():
    global sign1, sign2, pr1, pr2, res, text, f1
    if (f1 == 0):
        tb.setText(text)
    elif (f1 == 1):
        f1S()
    else: 
        f2S()
    clear()
    
    
    
    
def main():
    global sign1, sign2, pr1, pr2, res, text, tb, f1
    app = QApplication(sys.argv)
    window = QWidget()
    window.resize(420, 300)
    window.setWindowTitle("Primitive calculator")

    grid = QGridLayout(window)

    tb = QLabel(text)
    
    bt1 = QPushButton(window)
    bt1.setText("1")
    bt1.move(20, 200)
    bt1.setFixedSize(80, 40)
    bt1.show()
    bt2 = QPushButton(window)
    bt2.setText("2")
    bt2.move(120, 200)
    bt2.setFixedSize(80, 40)
    bt2.show()
    bt3 = QPushButton(window)
    bt3.setText("3")
    bt3.move(220, 200)
    bt3.setFixedSize(80, 40)
    bt3.show()
    bt4 = QPushButton(window)
    bt4.setText("4")
    bt4.move(20, 150)
    bt4.setFixedSize(80, 40)
    bt4.show()
    bt5 = QPushButton(window)
    bt5.setText("5")
    bt5.move(120, 150)
    bt5.setFixedSize(80, 40)
    bt5.show()
    bt6 = QPushButton(window)
    bt6.setText("6")
    bt6.move(220, 150)
    bt6.setFixedSize(80, 40)
    bt6.show()
    bt7 = QPushButton(window)
    bt7.setText("7")
    bt7.move(20, 100)
    bt7.setFixedSize(80, 40)
    bt7.show()
    bt8 = QPushButton(window)
    bt8.setText("8")
    bt8.move(120, 100)
    bt8.setFixedSize(80, 40)
    bt8.show()
    bt9 = QPushButton(window)
    bt9.setText("9")
    bt9.move(220, 100)
    bt9.setFixedSize(80, 40)
    bt9.show()
    bt0 = QPushButton(window)
    bt0.setText("0")
    bt0.move(120, 250)
    bt0.setFixedSize(80, 40)
    bt_plus = QPushButton(window)
    bt_plus.setText("+")
    bt_plus.move(320, 100)
    bt_plus.setFixedSize(80, 40)
    bt_plus.show()
    bt_minus = QPushButton(window)
    bt_minus.setText("-")
    bt_minus.move(320, 150)
    bt_minus.setFixedSize(80, 40)
    bt_minus.show()
    bt_multiply = QPushButton(window)
    bt_multiply.setText("*")
    bt_multiply.move(320, 200)
    bt_multiply.setFixedSize(80, 40)
    bt_multiply.show()
    bt_divide = QPushButton(window)
    bt_divide.setText("/")
    bt_divide.move(320, 250)
    bt_divide.setFixedSize(80, 40)
    bt_divide.show()
    bt_equals = QPushButton(window)
    bt_equals.setText("=")
    bt_equals.move(220, 250)
    bt_equals.setFixedSize(80, 40)
    bt_equals.show()
    bt_clear = QPushButton(window)
    bt_clear.setText("C")
    bt_clear.move(20, 250)
    bt_clear.setFixedSize(80, 40)
    bt_clear.show()

    tb = QLineEdit(window)
    tb.move(20, 50)
    tb.resize(380, 40)
    tb.setReadOnly(True)
    tb.setAlignment(Qt.AlignRight)
    tb.show()

    bt1.clicked.connect(w1)
    bt2.clicked.connect(w2)
    bt3.clicked.connect(w3)
    bt4.clicked.connect(w4)
    bt5.clicked.connect(w5)
    bt6.clicked.connect(w6)
    bt7.clicked.connect(w7)
    bt8.clicked.connect(w8)
    bt9.clicked.connect(w9)
    bt0.clicked.connect(w0)
    bt_plus.clicked.connect(w_plus)
    bt_minus.clicked.connect(w_minus)
    bt_multiply.clicked.connect(w_multiply)
    bt_divide.clicked.connect(w_divide)
    bt_equals.clicked.connect(w_equals)
    bt_clear.clicked.connect(w_clear)
    window.show()
    sys.exit(app.exec_())

if __name__ == "__main__":
    main()
```
